Реализовать функцию, принимающую два числа (позиционные аргументы) и выполняющую их деление. Числа запрашивать у пользователя, предусмотреть обработку ситуации деления на ноль.

```
def f_div(el1, el2):
    """ Функция делит el1  на  el2"""
    try:
        return el1/el2
    except ZeroDivisionError:
        print('Деление на ноль это ошибка')
```

```
print(f_div(int(input("Введите значение числителя")), int(input("Введите значение знаменателя"))))
```
